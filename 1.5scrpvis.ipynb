```python
%%time
text = TextBlob(str(filtered))
```

```python
tags_list = text.tags
```

```python
tags_list
```

```python
df_text = pd.DataFrame(tags_list)
df_text.columns = [ 'Words', "Word type"]
df_t=df_text.groupby('Word type').count().reset_index()
top20=df_t.nlargest(20, 'Words')
```

```python
plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 20):
      sns.barplot(x = "Words", y = "Word type",
     saturation = 0.9, data = top20).set_title("20th century key events - top 20 word types used")
```

```python
df = df_text[(df_text['Word type'] == "NN") | (df_text['Word type'] == "NNS") | (df_text['Word type'] == "NNS")]
df.columns = ["Word", "Occurrences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurrences'], ascending=False)
top15 = y.nlargest(15, 'Occurrences') 
```

```python
plt.figure(figsize=(19, 5))
with sns.dark_palette("xkcd:blue", 15):  #####!!!##!!!!! Here in the exercise was sns.dark.palette !
      sns.barplot(x="Word", y="Occurrences", 
      saturation=0.9, data = top15).set_title("Most frequently used nouns")
```

```python
verbs = [word for word, tag in tags_list if tag.startswith("VB")]
adjectives = [word for word, tag in tags_list if tag.startswith("JJ")]

verb_freq = Counter(verbs).most_common(15)
adj_freq = Counter(adjectives).most_common(15)

def plot_bar(data, title):
    df = pd.DataFrame(data, columns=['Word', 'Count'])
    plt.figure(figsize=(8, 5))
    sns.barplot(x='Word', y='Count', data=df)
    plt.xticks(rotation=45)
    plt.title(title)
    plt.show()

plot_bar(verb_freq, "Top 15 Verbs")
plot_bar(adj_freq, "Top 15 Adjectives")    
```
