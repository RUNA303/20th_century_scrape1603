```python
import pandas as pd
import time
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import os
import requests
from bs4 import BeautifulSoup
```

```python
# Automatically download the correct version of ChromeDriver
service = Service(ChromeDriverManager().install())

# Set Chrome options (optional but recommended)
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode (no UI)

# Launch the browser
driver = webdriver.Chrome(service=service, options=chrome_options)
```

```python
# Open the Wikipedia page
page_url = "https://en.wikipedia.org/wiki/Timeline_of_the_20th_century"
driver.get(page_url)

# Wait for the page to load
time.sleep(2)  
```

```python
# Get the entire page source
html_content = driver.page_source

# Use BeautifulSoup to parse the content
soup = BeautifulSoup(html_content, "html.parser")

# Extract text from the page
text_content = soup.get_text()

# Print a preview
print(text_content[:1000])  # Show the first 1000 characters
```

```python
# Define file path
output_file = "20th_century_scrape.txt"

# Save to file
with open(output_file, "w", encoding="utf-8") as file:
    file.write(text_content)

print(f"Scraped content saved to {output_file}")
```
